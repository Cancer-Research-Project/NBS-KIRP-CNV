## Cophenetic Correlation Coefficient 

```r
#install.packages('NMF')
library(NMF)
```

```r
cc_path = "./PCNet/kirc_combined_mad_binary_PCNetc3_i1000/PCNet_mad_c3_i1000_cc_matrix.csv"
cc.df = read.table(cc_path, header=TRUE, sep=",")
```

```r
head(cc.df)
```

```r
cc.mat = data.matrix(cc.df[-1])
head(cc.mat)
```

```r
ccc=cophcor(cc.mat)
write.csv(ccc,'./PCNet/kirc_combined_mad_binary_PCNetc3_i1000/coph.txt')
```

## Proportion of Ambiguous Clustering (PAC)

```r
#install.packages('diceR')
library(diceR)
pac=PAC(cc.mat)
write.csv(pac,'./PCNet/kirc_combined_mad_binary_PCNetc3_i1000/pac.txt')
```
